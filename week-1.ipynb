And after looking at every table, it turns out that pandas doesn't even see this matchups table as a table. I'll use beautiful soup to parse it instead.

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
print(driver.page_source)
```

```python
for i in soup.find_all(name="section", attrs={"class":"No-m No-pend Submod Selected"}):
    print(i.text)
```

```python
matchups_df = pd.DataFrame()
```

```python
text_only = ''
for i in soup.find_all(name="section", attrs={"class":"No-m No-pend Submod Selected"}):
    #print(i.text)
    #matchups_df.loc[len(matchups_df), "columnname"] = i.text
    value = i.text.replace("\n", "")
    value = value.replace(" ", "")
    #if not i.text.isspace():
    #    text_only += i.text
    matchups_df.loc[len(matchups_df), "column"] = value
```
